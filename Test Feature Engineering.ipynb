### Missing Values

```python
# lets handle categorical missing feature
cat_nan = [feature for feature in tdataset.columns if tdataset[feature].isnull().sum()>1 and tdataset[feature].dtype=='O']

for feature in cat_nan:
    print('{}:{}% missing values'.format(feature, np.round(tdataset[feature].isnull().mean(), 4)))
```

```python
def replace_cat_feature(tdataset, cat_nan):
    data = tdataset.copy()
    data[cat_nan]=data[cat_nan].fillna('Missing')
    return data

replace_cat_feature(tdataset,cat_nan)
```

```python
# now handle numerical missing values
numerical_nan = [feature for feature in tdataset.columns if tdataset[feature].isnull().sum()>1 and tdataset[feature].dtype!='O']

for feature in numerical_nan:
    print('{}:{}%missing values'.format(feature, np.round(tdataset[feature].mean(), 4)))
```

```python
# Replacing the numerical missing values
#creating new feature to capture Nan values
for feature in numerical_nan:
    median_value = tdataset[feature].median()
    tdataset[feature+'nan']= np.where(tdataset[feature].isnull(),1,0)
    tdataset[feature].fillna(median_value,inplace=True)
```

```python
tdataset[numerical_nan].isnull().sum()
```

```python
## Temporal Variables (Date Time Variables)

for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    tdataset[feature] = tdataset['YrSold']-tdataset[feature]
```

```python
tdataset[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']].head()
```

```python
num_features = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea']
for feature in num_features:
    tdataset[feature]=np.log(tdataset[feature])
```

```python
tdataset.head(2)
```

## Handling Rare Categorical Feature

```python
categorical_feature = [feature for feature in tdataset.columns if tdataset[feature].dtype=='O']
categorical_feature
```

## Handling rare categorical features

```python
for feature in categorical_feature:
    temp = tdataset[feature].value_counts()/ tdataset.shape[0]
    temp_df=temp[temp>0.01].index
    tdataset[feature]=np.where(tdataset[feature].isin(temp_df),tdataset[feature],'Rare_var')
```

```python
tdataset.head(10)
```

```python
#changing categorical values into numbers using label encoder
from sklearn.preprocessing import LabelEncoder
for label in categorical_feature:
    le = LabelEncoder()   
    tdataset[label]= le.fit_transform(tdataset[label])
```

## Feature Scaling

```python
feature_scale=[feature for feature in tdataset.columns if feature not in ['Id']]

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(tdataset[feature_scale])
```

```python
scaler.transform(tdataset[feature_scale])
```

```python
# transform the train and test set, and add on the Id variables
data = pd.concat([tdataset[['Id']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(tdataset[feature_scale]), columns=feature_scale)],
                    axis=1)
```
